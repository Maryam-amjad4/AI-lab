```python
import random

environment_rooms = {"A": random.choice(["Clean", "Dirty"]),
               "B": random.choice(["Clean", "Dirty"])}

model = {"A": "Unknown", "B": "Unknown"}
location = "A"
time_steps = 10

print("Initial Environment:", environment_rooms)
print(" ")

for t in range(time_steps):
    print(f"Time Step {t + 1}")
    percept = environment_rooms[location]
    print(f"Agent is in Room {location} and perceives it as {percept}")
    model[location] = percept

    if model[location] == "Dirty":
        print("Action: CLEAN")
        environment_rooms[location] = "Clean"
        model[location] = "Clean"
    else:
        location = "B" if location == "A" else "A"
        print("Action: Move to Room", location)

    print("Internal Model:", model)
    print("Actual Environment:", environment_rooms)
    print(" ")
```

```python
routes = {
    "Route A": {"distance": 10, "time": 15, "risk": "Low"},
    "Route B": {"distance": 8, "time": 20, "risk": "Medium"},
    "Route C": {"distance": 12, "time": 10, "risk": "High"}
}

risk_values = {"Low": 1, "Medium": 2, "High": 3}

def utility(route):
    dist = route["distance"]
    time = route["time"]
    risk = risk_values[route["risk"]]
    return (1 / dist) + (1 / time) + (1 / risk)

utilities = {}
route_names = ["Route A", "Route B", "Route C"]

for r in route_names:
    utilities[r] = utility(routes[r])


best_route = max(route_names, key=lambda r: utilities[r])

print("Utility Scores for Each Route:")
for r in route_names:
    print(r, ":", round(utilities[r], 3))

print("\nSelected Route:", best_route)
```
